# Importing Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# Loading Datasets

```python
data=pd.read_csv('Airbnb_datasets.csv',encoding_errors='ignore')
```

## Data Cleaning

```python
data.isnull().sum()

data.dropna(inplace=True)

data.isnull().sum()
```

```python
data.duplicated().sum()

data[data.duplicated()]

data.drop_duplicates(inplace=True)
```

```python

data.dtypes

data['id']=data['id'].astype(str)


data['host_id']=data['host_id'].astype(str)

data.dtypes
```

## Data Analysis (EDA)

Univariate Analysis


```python
df=data[data['price'] < 1500]

sns.boxplot(data=df,x='price')
```

```python
# Price Distribution

data['price']
plt.Figure(figsize=(8,5))
sns.histplot(data=df, x='price',bins=100)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()

```

```python
df.columns
```

```python
# availability_365 Distribution

data['availability_365']
plt.Figure(figsize=(8,5))
sns.histplot(data=df, x='availability_365')
plt.title('availability_365 Distribution')
plt.ylabel('Frequency')
plt.show()

```

```python
data.dtypes
```

```python
# Average price per bed
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

## Feature Engineering

```python
df['price per bed'] = df['price']/df['beds']
df.head()
```

Bi Variable Analysis

```python
df.columns
```

```python
sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')
```

```python
plt.Figure(figsize=(8,5))
plt.title("Locality And Review Dependency")
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.show()
```

```python
df.dtypes
```

```python
sns.pairplot(data=df,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
```

```python
#Geographical Distributions of AirBnB Listing
plt.Figure(figsize=(10,10))
plt.title('Geographical Distributions of AirBnB Listing')
sns.scatterplot(data=df,x='longitude',y='latitude',hue='room_type')
plt.show()
```

```python
df.dtypes
```

```python
# Heat Map

corr=df[['latitude','longitude','price','minimum_nights',
         
         'number_of_reviews','reviews_per_month','availability_365','beds']].corr()

corr
plt.Figure(figsize=(10,10))
sns.heatmap(data=corr,annot=True)
```
