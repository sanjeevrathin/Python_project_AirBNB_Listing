```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Raj", "Anu", "Ravi"],
    "Salary": [50000, 60000, 45000]
})

def salary_level(salary):
    if salary >= 50000:
        return "High Salary"
    else:
        return "Low Salary"

df["Category"] = df["Salary"].apply(salary_level)

print(df)
```
